## Digital University

Digital University is an initiative that was originally spearheaded by the Cyber community and the team at BESPIN.  This program partners with commericial learning platforms such as Udemy, Pluralsight, and Udacity to provide Department of the Air Force military and civilian employees with free access to a number of their course catalogs.  By accessing the entirety of Udemy's course catalog through Digital University's interface, I was able to handpick lessons from a variety of commericial, off-the-shelf courses and construct a learning path that would teach manpower professionals the data skills they need to level up their capabilities.  The best part?  We were able to do all of this at no cost, with no registration limits, and the instructor overhead was limited to one person facilitating this part-time.



## Overview

At it's core, the Manpower Data Analytics Learning Path was an experiment.  We wanted to see if we can use free, accessible learning platforms to provide manpower analysts with more opportunities to broaden their data analytic skills.  The program was intentionally self-paced, hands-on with respect to a constructivist-inspired pedagogical framework, and hands-off with respect to instructor touch-time and active feedback.  The desired outcome of this program was familiarization, not mastery, with an emphasis on practical application.

Throughout this test, we created two courses: a 3-week course in Data Visualization which had a combined 230 participants and a 4-week course in Inferential Statistics which had 55 participants.

During today's presentation, I will focus on the Data Visualization course and share some of the projects, quantify changes we observed in student performance, and discuss trends in student feedback.  In the end, I will conclude with general lessons we learned from all the courses we ran that are widely applicable for any function seeking to embrace digital learning.



## Review Course Elements

Each course we developed was a collection of weekly modules, and each module contained four key attributes.  First, students were provided with  educational videos from Digital University and reading excerpts from AF eLearning.  Second, students were encouraged to complete an open-ended project, which would take the skills they learned and apply it to a relevant manpower context.  Third, digital learning was supplemented with weekly review sessions, where I would make myself available to meet with students over Microsoft Teams and answer any questions they may have regarding the material.  To account for personnel stationed around the world, these review sessions were held twice each week: once in the morning and once in the afternoon.  Lastly, we provided students with additional resources for further study should they be interested in learning more.  For example, if a student really enjoyed our Exploratory Data Analysis module, they could continue their studies with links we provided to free extensive courses provided by Harvard, MIT, and other universities through the edX platform as well as additional textbooks available through AF eLearning.



As I mentioned before, our Data Visualization was comprised of three week-long modules.  Each module contained a number of skills that we wanted students to learn, and each module culminated in an applied project.  The first week of this course introduced students to the practice of exploring large datasets.  The second week introduced students to the art of data visualization using Microsoft Excel.  In the last week, the students chose to specialize in either Tableau or Power BI and learned to construct interactive reports for decision markers.



## Share Course Projects

In the first week's project, students were provided with fictitious sign-in logs from a base Military Personnel Flight.  These files contained more that 23,000 records and students were challenged to explore the data and find insights regarding the office's operations that they could then present to a flight commander to better inform their day-to-day operations.



In the second week's project, students were tasked with pulling a unit manpower document applicable to their current workcenter and construct interactive dashboard in Microsoft Excel that could be used by decision makers to visualize the alignment of their current manpower.  Although students were not required to then provide these products to the decision makers, some students elected to do so and reported especially positive feedback.

## Quantify Student Growth

To evaluate the degree to which student performance improved during this program, students were provided surveys at the start and end of the course.  This data is self-reported by students and ranges from "I don't know what this is" at the bottom of our y-axis to "I could teach this topic to other" at the top of our y-axis.  

To best display the distribution of the class's knowledge, I've elected to use a box-and-whisker plot.  The x in each plot indicate the class's average level of perceived knowledge, the bottom of each box indicates the 25th percentile of the class, and the top of each box indicates the 75th percentile.

If student performance improved as a result of the course, we would see green indicators notably higher than the blue indicators, which we do in across all topics and skills.  While there were still some students who reported the same level of understanding as when they began the course, the preponderence of the class improved.  This is noted by the green x's and boxes being higher than the blue x's and boxes.



On this second set of topics, students continued to show improvements.  Notably, all three subjects showed stark increases with the 25th percentile of the class reaching the same level of proficiency that the 75th percentile of the class had at the start of the course.



Our starkest increases in performance were in relation to using Tableau and Power BI.  This is presumably due to many students having never been exposed to these software applications prior to this course.



## Discuss Student Feedback

In the student feedback, we received predominantly positive comments but also a number of constructive critiques.  In general, students tended to find the class to be informative and appreciated the flexibility of the course.  Many students commented on how the course provided options for visual learners, audio learners, and learners that prefer to learn by reading.  Some students also enjoyed being able to keep their own pace through the course, with some skipping videos that they were already familiar with and others getting to rewatch videos multiple times until they felt confident in the material.

Regarding the critiques we received, the most prominent were from students who prefer to learn in an in-person course.  Having to juggle online learning with work and home instead of getting to go TDY proved difficult for some students to manage.  Additionally, there were students who felt that the project examples they completed throughout the course were not relevent to the type of work they accomplish from day-to-day.  This criticism makes sense as the manpower enterprise is spread across all levels of command with various offices performing different types of work.



## Final Takeaways

To conclude, I'd like to leave you with the four biggest lessons we learned from running this program.

First, the digital learning environment works particularly well with less technical, more practical application topics.  We found that during the Data Visualization course, students tended to really enjoy having the opportunity to explore their open-ended projects at their own pace and in a direction of their choosing.

Second, the digital learning environment can work for more technical topics, but needs to be supplemented with more structure and instructor interaction.  Our Inferential Statistics course was also generally well-received, but we had longer review sessions and more students requesting one-on-one meetings with the instructor to review content.

Third, virtual courses can present a challenge when balancing regular workload and home life.  This takeway is perhaps emblematic of something many of us learned last year when we shifted to teleworking - striking a balance is important.  The supervisors of students should be aware of the level of coursework prior to the class beginning and attempt to the extent possible to be flexible with workload.

Lastly, teamwork was shown to present both challenges and opportunities.  We had so many students sign up for one of our courses that we decided to do an experiment wherein we split the class into three groups.  One group had students work individually on projects, another group had students work in pairs, and the last group had students work in teams of four.  In the end, we found that students working in pairs was the best balance, both maximizing the likelihood that students would complete the project as partners were holding each other accountable while also minimizing the struggling of scheduling time to meet with multiple teammates who may be in varying time zones.



All in all, the manpower enterprise found Digital University to be an enticing alternative to the traditional methods of Air and Space Force training.  By experimenting with a virtual learning program, we found a way to train hundreds of manpower professionals without needing to invest in course materials, TDY expenses, or incur long absences from work.  While it had it's strengths and weaknesses, I would recommend it as an option for any functional community seeking to extend opportunities for their Airmen and Guardians to learn.
